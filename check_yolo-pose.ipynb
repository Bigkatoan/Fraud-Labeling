```python
camera = cv2.VideoCapture(0)

while True:
    ret, frame = camera.read()

    result = model(frame, save = False)
    result = result[0].plot()
    cv2.imshow(' Camera', result)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
